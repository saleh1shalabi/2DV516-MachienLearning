### The implementation of bkmeans
I chooesed to not use the kmeans from sklearn but to implemente it on my own wich seems to work as it should

```python
def find_cent(X):
    """
    findes the centorids of the cluster
    """
    s = np.sum(X, axis=0)
    s = s / X.shape[0]
    return s

def euclidean_distance_bk(X, point):
    return np.linalg.norm(X - point)

def sse(X, point):
    """
    sum of squared error
    """
    return euclidean_distance_bk(X, point) ** 2

def kmeans(X, iter, labels = None):
    
    """
        kmenas function
        takes the array to cluster, the labels is what labels they should take
        
        without labels it gives 0, 1 as clusters
        
        
    
    """
    
    SSEs = []

    y = np.zeros(X.shape[0], dtype=np.int32)
    ones = np.random.choice(X.shape[0], round(X.shape[0]/2), replace=False)
    y[ones] = 1
    zeros = np.where(y == 0)

    centroids = [find_cent(X[zeros]), find_cent(X[ones])]

    g = sse(X[zeros],centroids[0]) + sse(X[ones], centroids[1])
    SSEs.append(g)

    for i in range(iter):

        for j in range(X.shape[0]):
            if euclidean_distance_bk(X[j], centroids[0]) < euclidean_distance_bk(X[j], centroids[1]):
                y[j] = 0
            elif euclidean_distance_bk(X[j], centroids[0]) > euclidean_distance_bk(X[j], centroids[1]):
                y[j] = 1
            else: pass

        ## some maight not cahnge so we should look up all
        ones = np.where(y == 1)
        zeros = np.where(y == 0)

        centroids = [find_cent(X[zeros]), find_cent(X[ones])]

        g = sse(X[zeros], centroids[0])
        g += sse(X[ones], centroids[1])

        if g == SSEs[len(SSEs) - 1] and g == SSEs[len(SSEs) - 2]:
            break
        else: SSEs.append(g)

    if labels != None and len(labels) == 2:
        zeros = np.where(y == 0)
        ones = np.where(y == 1)
        y[zeros] = labels[0]
        y[ones] = labels[1]
    return y

def bkmeans(X, K, all = False, max_iter = 100):
    
    """
    the bkmeans function 
    
    takes  X: the array to cluster 
           K: the number of clusters
           
        all if True the function returns a list of each step of the clustring process
        
        max_iter is how many times the kmenas function should iterate
        
        
    
    """
    
    
    
    if K == 0:
        raise ValueError("K must be larger than 0")
    if K == 1:
        return np.zeros(X.shape[0])
    st = []
    y = np.zeros(X.shape[0])
    st.append(y)


    ## since we know that all is one value we run the first iter outside for efficency
    y = kmeans(X, max_iter, [0, 1])
    n = np.copy(y)
    st.append(n)

    clusters = 2

    while clusters != K:
        sums = []
        vals = []

        for i in np.unique(y):

            sums.append(np.sum(np.where(y == i)))
            vals.append(i)

        s = max(sums)
        s = sums.index(s)

        y_to_bi = np.where(y == vals[s])
        y[y_to_bi] = kmeans(X[y_to_bi], max_iter, [y[y_to_bi][0], clusters])
        n = np.copy(y)
        st.append(n)
        clusters += 1


    if all: return st
    return y
```

### Sammon mapping

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist
from sklearn.datasets import make_blobs, make_s_curve, make_swiss_roll
from threading import Thread
```

```python

def euclidean_distance(X, array):
    """ 
    euclidean distance between 2 arrays of same shape point to point
    """
    return np.linalg.norm(X - array, axis=1)


def array_distances(X):
    
    """
    calculates all distances between each and every point in array
    """

    return pdist(X, "euclidean") # faster than my implementation :(

    # good when X sampels not larger than 1000
    # dists = np.array([])
    # i = 1
    # for i in range(X.shape[0]):
    #   j = X[i+1:]
    #   z = np.full(j.shape, X[i])
    #   o = euclidean_distance(z, j)
    #   dists = np.append(dists, o)
    #   i+=1
    # return dists



def sammon_stress(d_dists, real_dists):
    
    """
        calculates the E factor of the input dim and output dim
    """
    d_dists[d_dists < 1e-10] = 1e-7
    real_dists[real_dists < 1e-10] = 1e-7

    r = np.sum(real_dists)

    delta_dists = (d_dists - real_dists) ** 2
    diff = np.nan_to_num(delta_dists / real_dists)

    E = (1/r) * sum(diff)

    return E





def best_params(X):
    """
        was not required in the assignment but I used it to test for my self

        it find the axsis that have least sammon stress E and returns them
    """


    dists_X = array_distances(X)
    b_f = {}

    for i in range(X.shape[1]):
        l = {}
    for j in range(X.shape[1]):
      if i != j:
        y = X[:,[i,j]]
        dists_y = array_distances(y)
        l[j] = sammon_stress(dists_X, dists_y)

    b_f[i] = l

    least = []
    for k, v in b_f.items():
        least.append(min(v.values()))

    # print(min(least))

    params = []

    for k, v in b_f.items():
        for i, j in v.items():
          if j == min(least):
            params.append(k)
            params.append(i)


    return list(set(params))







def gradient(X, y , y_new, i, alpha, c):
    """
        method used to run multi threads
        it calculate the new y values
    """

    Xj = np.delete(np.copy(X), i, axis=0)
    yj = np.delete(np.copy(y), i, axis=0)
    Xi = np.full(Xj.shape, X[i])
    yi = np.full(yj.shape, y[i])

    xij_dists = euclidean_distance(Xi, Xj).reshape(Xi.shape[0], 1)

    xij_dists[xij_dists< 1e-10] = 1e-10

    yij_dists = euclidean_distance(yi, yj).reshape(Xi.shape[0], 1)  # reshape
    yij_dists[yij_dists< 1e-10] = 1e-10

    yi_diff_yj = yi - yj

    xy_diff = xij_dists - yij_dists

    denom = xij_dists * yij_dists
    denom[denom < 1e-10] = 1e-10

    p1 = np.nan_to_num(-2/c) * np.sum(np.nan_to_num(xy_diff/denom) * yi_diff_yj)

    t = 1 + np.nan_to_num(xy_diff / yij_dists)

    dd = np.nan_to_num( (yi_diff_yj ** 2) / yij_dists )
    p2 = np.nan_to_num(-2/c) * np.sum(np.nan_to_num(1/denom) * (xy_diff - (dd*t)))
    delta = p1 / np.abs(p2)
    y_new[i] = y[i] - alpha * delta

    
    
    
    
def sammon(X, alpha, iter, error, y_s = 2, chunk_size = 5, best_feat = False):
    
    """
        all of y_s, chunk_size and beast_feat params was not required
        y_s: is the output space that will be retuerned. 

        chunk_size: is the number of threads to be used. 

                    using high number of threads when the array have large size
                    will overflow the memory, (use wisly what deppending on the array size)


        best_feat: takes the best axis and returns 2 dim array 


    !!The method returns the output space and the axises it shows!!.

    """


    columns = []

    if best_feat:
        columns = best_params(X)



    # since the random choosing of the axises the output space will look diffrent each time.
    else:
        for i in range(y_s):
          rand_col = random.randint(0, X.shape[1]-1)
          while rand_col in columns:   ### make sure to not choose the same axis 2 times.
            rand_col = random.randint(0, X.shape[1]-1)
          columns.append(rand_col)

    y = X[:, columns]
    dists_x = array_distances(X)
    #   print("array calcs")

    c = np.sum(dists_x)

    y_new = np.copy(y)

    E_old = 100 # just a start value

    for n in range(iter):

        if n == 0:
          dists_y = array_distances(y)
          E_old = sammon_stress(dists_y, dists_x)
          E = E_old

        else:
          dists_y = array_distances(y_new)
          E = sammon_stress(dists_y, dists_x)


    ## we look at the E if it is smaller than the error
    ## we should even look if it start to increase, then we are goinge wrong direction
    ## it accepts as much as 2 times the error
    ## when taken abs(E_old - E) if we then instade of increasing, decreasing then it will break 
    ## thats why we take the negative value of it and look agains the positive value
    if E < error or ( abs(E_old - E) > error*2 and -(E_old - E) > (E_old - E)) : 
    #       print(E)
        return y, columns

    else:
      y = np.copy(y_new)
      E_old = E


        ## uncomment to see the proces of the function
    #     print("iter:", n, "E:", E)

    ts = []
    for i in range(X.shape[0]):
        if len(ts) != chunk_size:
            t = Thread(target=gradient, args=(X, y, y_new, i, alpha, c))
            t.start()
            ts.append(t)
        else:
            for t in ts:
                t.join()
            ts = []

    return y, columns
```

#### Normalization function
I noticed that for some of the datasets, normalaizing gives better result.

```python

def normalize_values(X):
    
    """
        normelaize values of array

    """
    mean_values = []
    std_values = []
    for i in range(X.shape[1]):
      mean_values.append(np.mean(X[:, i]))
      std_values.append(np.std(X[:, i]))

    for i in std_values:
        if i < 1e-8 : i = 1e-5




    Xn = []
    for i in range(X.shape[1]):
      g = np.nan_to_num((X[:, i] - mean_values[i]) / std_values[i])
      Xn.append(np.array(g).T)
    Xn = np.array(Xn).T
    return Xn
```
